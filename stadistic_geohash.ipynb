```python
import os
import pandas as pd
from collections import Counter
```

```python
def check_path_exists(path):
    print(os.path.exists(path))
```

```python
precision = [6,7,8]
p = precision[0]
trajectory_id = "001_0"
```

```python
PATH_TRAJECTORIES_GEOHASH = f"D:/Topicos en manejo de grandes volumenes de datos/Proyecto/output_geohash/precision_{p}"
check_path_exists(PATH_TRAJECTORIES_GEOHASH)

format_plt = ".plt"
format_csv = ".csv"


FINAL_PATH_TRAYECTORY_GEOHASH = PATH_TRAJECTORIES_GEOHASH + "/" +trajectory_id + format_csv
print(FINAL_PATH_TRAYECTORY_GEOHASH)

check_path_exists(FINAL_PATH_TRAYECTORY_GEOHASH)

df_2 = pd.read_csv(FINAL_PATH_TRAYECTORY_GEOHASH)
print(df_2.head())
print(df_2.shape)
```

```python
# cuantas veces aparece cada geohash
geohash_counts = df_2['geohash'].value_counts()
print(geohash_counts)
geohash_counts.mean()
```

```python
def mean_points_per_geohash_streaming(base_path, sample_fraction=1.0):
    geohash_counter = Counter()
    files = [f for f in os.listdir(base_path) if f.endswith(".csv")]
    print(base_path)
    print(f"Total de archivos encontrados: {len(files)}")

    if sample_fraction < 1.0:
        import random
        k = int(len(files) * sample_fraction)
        files = random.sample(files, k)
        print(f"Usando {k} archivos de {len(files)}")

    for file in files:
        file_path = os.path.join(base_path, file)

        try:
            # Lee solo la columna "geohash"
            for chunk in pd.read_csv(file_path, usecols=["geohash"], chunksize=50_000):
                geohash_counter.update(chunk["geohash"])
        except Exception as e:
            print(f"Error en archivo {file}: {e}")

    total_points = sum(geohash_counter.values())
    total_geohashes = len(geohash_counter)

    media = total_points / total_geohashes if total_geohashes > 0 else 0
    std = pd.Series(list(geohash_counter.values())).std()
    median = pd.Series(list(geohash_counter.values())).median()
    return media, geohash_counter, std, median, min(geohash_counter.values()), max(geohash_counter.values())
```

```python
precision = [6,7,8]
p = precision[0]

PATH_TRAJECTORIES_GEOHASH = f"D:/Topicos en manejo de grandes volumenes de datos/Proyecto/output_geohash/precision_{p}"
check_path_exists(PATH_TRAJECTORIES_GEOHASH)
base = PATH_TRAJECTORIES_GEOHASH
mean, counter, std, median, min_count, max_count = mean_points_per_geohash_streaming(base, sample_fraction=1.0)
print(f"Media de puntos por geohash: {mean}")
print(f"Número total de geohashes únicos: {len(counter)}")
print(f"Desviación estándar: {std}")
print(f"Mediana de puntos por geohash: {median}")
print(f"Mínimo de puntos por geohash: {min_count}")
print(f"Máximo de puntos por geohash: {max_count}")
```
