```python
import os
import pandas as pd
from collections import Counter
import random
```

```python
def seleccionar_porcentaje(carpeta, percent, random_seed=42):

    random.seed(random_seed)

    archivos = [os.path.join(carpeta, f) 
                for f in os.listdir(carpeta) 
                if os.path.isfile(os.path.join(carpeta, f))]

    total = len(archivos)
    k = int(total * percent)

    return random.sample(archivos, k)
```

```python
def load_to_vector_dfs(file_list):
    vector_dfs = []
    for file in file_list:
        df = pd.read_csv(file)
        vector_dfs.append(df)
    return vector_dfs
```

```python
def lcss_geohash(seq1, seq2):

    if len(seq1) < len(seq2):
        seq1, seq2 = seq2, seq1  

    prev = [0] * (len(seq2) + 1)
    curr = [0] * (len(seq2) + 1)

    for i in range(1, len(seq1) + 1):
        for j in range(1, len(seq2) + 1):
            if seq1[i-1] == seq2[j-1]:
                curr[j] = prev[j-1] + 1
            else:
                curr[j] = max(prev[j], curr[j-1])
        prev, curr = curr, prev 

    return prev[-1]
```

```python
def df_to_sequence(df):
    return df["geohash"].astype(str).tolist()
```

```python
import random

def compute_lcss_sample(dfs, num_pairs=20, random_seed=42):
    random.seed(random_seed)
    results = []

    sequences = [df_to_sequence(df) for df in dfs]

    for _ in range(num_pairs):
        a, b = random.sample(sequences, 2)
        score = lcss_geohash(a, b)
        results.append(score)

    return results
```

```python
lcss_scores = compute_lcss_sample(vector_dfs, num_pairs=10000, random_seed=42)
print(lcss_scores)
print("Media LCSS:", sum(lcss_scores)/len(lcss_scores))
```

```python
def read_first_line_txt(txt_path):

    with open(txt_path, "r") as f:
        line = f.readline().strip()  # lee la primera lnea y elimina espacios/newlines

    # Quitar los corchetes si existen
    line = line.strip("[]")

    # Separar por comas y convertir a enteros
    array = [int(x) for x in line.split(",") if x.strip() != ""]
    return array

```

```python
lcss_p6 = read_first_line_txt("p6_lcss.txt")
lcss_p7 = read_first_line_txt("p7_lcss.txt")
lcss_p8 = read_first_line_txt("p8_lcss.txt")
print(len(lcss_p6))
print(len(lcss_p7))
print(len(lcss_p8))
```

```python
def frequency_of_number(arr, num):
    return Counter(arr)[num]

def count_greater_than(arr, num):
    return sum(1 for x in arr if x > num)


arr = [1, 5, 2, 8, 3, 7]
print(count_greater_than(arr, 4))  # 3 (5, 8, 7)
```
