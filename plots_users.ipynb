```python
import os
from collections import defaultdict

def contar_trayectorias_por_user(directory):
    counts = defaultdict(int)

    for file in os.listdir(directory):
        if file.endswith(".plt") and "_" in file:
            user_id = file.split("_")[0]  # lo que va antes del primer "_"
            counts[user_id] += 1

    return dict(counts)
```

```python
counts = contar_trayectorias_por_user("../output_renombrado")
```

```python
counts
```

```python
data = counts
```

```python
labels = list(data.keys())
values = list(data.values())

threshold = sum(values) * 0.02   # categoras menores al 2% del total
labels_big = []
values_big = []
otros_total = 0

for label, value in data.items():
    if value < threshold:
        otros_total += value
    else:
        labels_big.append(label)
        values_big.append(value)


if otros_total > 0:
    labels_big.append("Otros")
    values_big.append(otros_total)

plt.figure(figsize=(10, 10))
plt.pie(values_big, labels=labels_big, autopct='%1.1f%%')
plt.title("Distribucin de categoras")
plt.savefig("distribucion_categorias.png", dpi=300)
plt.show()
```
