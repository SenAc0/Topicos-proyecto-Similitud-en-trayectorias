```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import os
from pathlib import Path
import folium
```

```python
def seleccionar_porcentaje(carpeta, percent, random_seed=42):

    random.seed(random_seed)

    archivos = [os.path.join(carpeta, f) 
                for f in os.listdir(carpeta) 
                if os.path.isfile(os.path.join(carpeta, f))]

    total = len(archivos)
    k = int(total * percent)

    return random.sample(archivos, k)
```

```python
selected_files = seleccionar_porcentaje('../output_renombrado', 1, random_seed=42)

print(len(selected_files))
print(selected_files)
```

```python
def cargar_trayectoria_plt(path):
    coords = []
    with open(path, "r") as f:
        for i, line in enumerate(f):
            if i < 6:
                continue  # saltar header
            parts = line.strip().split(',')
            lat = float(parts[0])
            lon = float(parts[1])
            coords.append((lat, lon))
    return coords
```

```python
def mapa_con_trayectorias(lista_rutas):
    # Obtener coordenadas iniciales
    first_coords = cargar_trayectoria_plt(lista_rutas[0])

    m = folium.Map(
        location=first_coords[0],
        zoom_start=13,
        tiles="https://tile.openstreetmap.org/{z}/{x}/{y}.png",
        attr="OpenStreetMap"
    )

    for ruta in lista_rutas:
        coords = cargar_trayectoria_plt(ruta)
        folium.PolyLine(coords, weight=3, opacity=0.7).add_to(m)

    m.save("mapa_output.html")

    return m
```

```python
import folium
import colorsys

def generar_colores(n):
    colores = []
    for i in range(n):
        h = i / n
        r, g, b = colorsys.hsv_to_rgb(h, 0.85, 0.85)
        colores.append('#%02x%02x%02x' % (int(r*255), int(g*255), int(b*255)))
    return colores


def mapa_con_trayectorias_colores(lista_rutas):
    first_coords = cargar_trayectoria_plt(lista_rutas[0])

    m = folium.Map(
        location=first_coords[0],
        zoom_start=13,
        tiles="CartoDB positron"   # mapa simlep
    )

    colores = generar_colores(len(lista_rutas))

    for i, ruta in enumerate(lista_rutas):
        coords = cargar_trayectoria_plt(ruta)
        folium.PolyLine(
            coords,
            color=colores[i],
            weight=3,
            opacity=0.8
        ).add_to(m)

    m.save(f"mapa_output_{len(lista_rutas)}.html")
    return m

```
